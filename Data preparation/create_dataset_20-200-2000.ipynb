Impostare la lunghezza delle sequenze

```python
max_seq_lenght = [20,200,2000]
```

```python
class enhancer:
    def __init__(self, gene, sequence):
        self.gene = gene
        self.sequence = sequence
    def __str__(self):
        return "Gene: "+ str(self.gene) +" sequence: "+ str(self.sequence)+"\n"
```

```python
PATH1 = "/content/drive/MyDrive/Data/enhancer.txt"

import re

mean = 0.0

with open(PATH1,"r") as file_enhancer:
    list_enhancer = []
    for line in file_enhancer:
      gene, sequence = line.split('\t')
      tmp = enhancer(gene, sequence)
      list_enhancer.append(tmp)
      mean += len(sequence)

print(len(list_enhancer))
mean = mean / len(list_enhancer)
```

```python
all_mins = []

for max in max_seq_lenght:
  mins=[]
  if max == 20:
    for item in list_enhancer:
      if len(item.sequence) < 2850:
        mins.append(item)
  elif max== 200:
    for item in list_enhancer:
      if len(item.sequence) < mean*9:
        mins.append(item)
  elif max == 2000:
    mins = list_enhancer
  all_mins.append(mins)
  print(f"{max}:{len(mins)}")
```

L'ultimo carattere non ci vuole nella sequenza

```python
list_enhancers=[]
for mins in all_mins:
  list_enhancer = mins
  for item in list_enhancer:
    item.sequence = item.sequence[:-1]
  list_enhancers.append(list_enhancer)

```

In questo pezzo di codice vengono tagliati gli enhancer a max_seq_lenght

```python
j=0
for list_enhancer in list_enhancers:
  i = 0
  while 1:
    dim=len(list_enhancer)
    if(i >= dim):
      break
    if (len(list_enhancer[i].sequence)> max_seq_lenght[j]):
      copy = list_enhancer[i].sequence
      tmp = enhancer(list_enhancer[i].gene, copy[max_seq_lenght[j]:])
      list_enhancer.append(tmp)
      list_enhancer[i].sequence = copy[0:max_seq_lenght[j]]

    if(len(list_enhancer[i].sequence)< max_seq_lenght[j]):
      copy = list_enhancer[i].sequence
      while (len(copy)< max_seq_lenght[j]):
        copy += 'N'
      list_enhancer[i].sequence = copy[0:max_seq_lenght[j]]
    i += 1
  j+=1
```
