Impostare la lunghezza delle sequenze

```python
max_seq_lenght = 1000
```

```python
class enhancer:
    def __init__(self, gene, sequence):
        self.gene = gene
        self.sequence = sequence
    def __str__(self):
        return "Gene: "+ str(self.gene) +" sequence: "+ str(self.sequence)+"\n"
```

```python
PATH1 = "/content/drive/MyDrive/Data/enhancer.txt"

import re

mean = 0.0

with open(PATH1,"r") as file_enhancer:
    list_enhancer = []
    for line in file_enhancer:
      gene, sequence = line.split('\t')
      tmp = enhancer(gene, sequence)
      list_enhancer.append(tmp)
      mean += len(sequence)

print(len(list_enhancer))
mean = mean / len(list_enhancer)
```

```python
mins = []
if max_seq_lenght == 100:
  for item in list_enhancer:
    if len(item.sequence) < mean:
      mins.append(item)
elif max_seq_lenght == 10:
  for item in list_enhancer:
    if len(item.sequence) < 2000:
      mins.append(item)
elif max_seq_lenght == 1000:
  mins = list_enhancer

print(len(mins))
```

L'ultimo carattere non ci vuole nella sequenza

```python
list_enhancer = mins
for item in list_enhancer:
  item.sequence = item.sequence[:-1]
```

In questo pezzo di codice vengono tagliati gli enhancer a max_seq_lenght

```python
i = 0

while 1:
  if(i >= len(list_enhancer)) :
    break
  if (len(list_enhancer[i].sequence)> max_seq_lenght):
    copy = list_enhancer[i].sequence
    tmp = enhancer(list_enhancer[i].gene, copy[max_seq_lenght:])
    list_enhancer.append(tmp)
    list_enhancer[i].sequence = copy[0:max_seq_lenght]

  if(len(list_enhancer[i].sequence)< max_seq_lenght):
    copy = list_enhancer[i].sequence
    while (len(copy)< max_seq_lenght):
      copy += 'N'
    list_enhancer[i].sequence = copy[0:max_seq_lenght]


  i += 1
```

```python
from Bio import SeqIO

promoter_list = []


# Apri il file .fa e leggi le sequenze
with open(promoter_path, "r") as handle:
    for record in SeqIO.parse(handle, "fasta"):
       id = record.id
       sequence = record.seq
       tmp = promoter(id, sequence)
       promoter_list.append(tmp)
```
